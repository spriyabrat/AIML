4. Write a function with name â€œGoal_diff_countâ€ which should return all the teams with their Goal Differences.
   Find the team which has maximum and minimum goal difference

```python
def Goal_diff_count(x):
    laliga.iloc[:,8:10] = laliga.iloc[:,8:10].astype(int)
    GoalsFor,GoalsAgainst=x
    return(GoalsFor-GoalsAgainst)
```

```python
laliga['Goal Difference'] = laliga[['GoalsFor','GoalsAgainst']].apply(Goal_diff_count,axis=1)
laliga.head()
```

```python
print('The team with Max Goal Difference:')
print(laliga['Team'].iloc[laliga['Goal Difference'].idxmax()])
print('The team with Min Goal Difference:')
print(laliga['Team'].iloc[laliga['Goal Difference'].idxmin()])
```
